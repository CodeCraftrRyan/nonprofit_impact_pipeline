```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
#  Load cleaned data
df = pd.read_csv("data/processed/cleaned_marketing_data.csv")
```

```python
# Select numeric features for clustering
features = [
    "age", "income", "total_spent",
    "numwebvisitsmonth", "numstorepurchases",
    "numwebpurchases"
]
X = df[features]
```

```python
#  Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Run KMeans
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X_scaled)
```

```python
#  Add cluster labels to original DataFrame
df["donor_segment"] = kmeans.labels_
```

```python
# Plot clusters in 2D (using 2 selected features)
plt.figure(figsize=(8, 6))
plt.scatter(X["income"], X["total_spent"], c=df["donor_segment"], cmap="viridis")
plt.xlabel("Income")
plt.ylabel("Total Spent")
plt.title("Donor Segments by Income and Total Spend")
plt.colorbar(label="Segment")
plt.show()
```
